```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/twitter_sentiment.csv', header=None, index_col=[0])
df.head() #show first five rows

```

```python
df = df[[2,3]].reset_index(drop=True) #index not starting from 1 and only 2 and 3 column is show
```

```python
 # df.columns = ['sentiment,  'text']
df.columns = ['sentiment', 'text']
df.head()
```

```python
df.info() #show basic infromation about the data
```

```python
df.isnull().sum() #show total null data
#some tweet have text null or value less words, so we have to drop those data
df.dropna(inplace=True) #drop all null values
```

```python
df['text'].apply(len).value_counts() # give length of every tweet , here some tweet have one length
sum(df['text'].apply(len)>5),sum(df['text'].apply(len)<=5)
#here 72883 means tweet having length greater then 5 and 2113 less tha or equal to 5
```

DATA CLEANING


```python
df['text']=df['text'].apply(lambda x: x.lower()) #convert to lower case 
df['text']=df['text'].apply(lambda x: ps.remove_urls(x)) #remove all urls
df['text']=df['text'].apply(lambda x: ps.remove_html_tags(x)) # removed alll html tags
df['text']=df['text'].apply(lambda x: ps.remove_rt(x)) # all retweeets are removed
df['text']=df['text'].apply(lambda x: ps.remove_special_chars(x)) # all special charcters removed

```

TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text'],df['sentiment'],test_size=0.2,random_state=0)
```

```python
X_train.shape , X_test.shape
```

MODEL BUILDING AND TRAINING  

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline #pipeline keep TfidfVectorizer and RandomForestClassifier together
```

```python
stopwords_list = list(stopwords)

clf =Pipeline([('tfid',TfidfVectorizer(stop_words=stopwords_list)),('rfc',RandomForestClassifier(n_jobs=-1))]) #n_jobs means all available core are used
#stopwords all not included
clf.fit(X_train,y_train)
```

EVALUATION

```python
from sklearn.metrics import accuracy_score

predictions = clf.predict(X_test)
print(accuracy_score(y_test, predictions)*100)
```
